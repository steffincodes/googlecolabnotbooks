```python
#@title
# list of available markers in matplotlib
MAT_MARKER = [
              [".", "point"],
              [",", "pixel"],
              ["o", "circle"],
              ["v", "triangle_down"],
              ["^", "triangle_up"],
              ["<", "triangle_left"],
              [">", "triangle_right"],
              ["1", "tri_down"],
              ["2", "tri_up"],
              ["3", "tri_left"],
              ["4", "tri_right"],
              ["8", "octagon"],
              ["s", "square"],
              ["p", "pentagon"],
              ["P", "plus (filled)"],
              ["*", "star"],
              ["h", "hexagon1"],
              ["H", "hexagon2"],
              ["+", "plus"],
              ["x", "x"],
              ["X", "x (filled)"],
              ["D", "diamond"],
              ["d", "thin_diamond"],
              ["|", "vline"],
              ["_", "hline"],
              [0, "tickleft"],
              [1, "tickright"],
              [2, "tickup"],
              [3, "tickdown"],
              [4, "caretleft"],
              [5, "caretright"],
              [6, "caretup"],
              [7, "caretdown"],
              [8, "caretleft (centered at base)"],
              [9, "caretright (centered at base)"],
              [10, "caretup (centered at base)"],
              [11, "caretdown (centered at base)"],
              ["None", "nothing"],
              ["$sc$", "Render the string using mathtext."]
]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

y=np.random.rand(10,1)
nplots=len(MAT_MARKER)
ncols,ndims=6,3

fig,ax=plt.subplots(
    nrows=math.ceil(nplots/ncols),
    ncols=ncols,
    figsize=(ndims*ncols,
             ndims*nplots/ncols),
    sharex=True
)

for i,m in enumerate(MAT_MARKER):
  ax[int(i/ncols)][i%ncols].plot(
      y,'o',
      marker=m[0]
  )
  ax[int(i/ncols)][i%ncols].set_title(str(m[0])+":"+m[1])
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,100)
fig,ax = plt.subplots()

ax.plot(x,x,label="linear")
ax.plot(x,x**2,label="quadratic")
ax.plot(x,x**3,label="cubic")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_title("Simple Plots")
ax.legend();
```
