```python
#@title
# list of available markers in matplotlib
MAT_MARKER = [
              [".", "point"],
              [",", "pixel"],
              ["o", "circle"],
              ["v", "triangle_down"],
              ["^", "triangle_up"],
              ["<", "triangle_left"],
              [">", "triangle_right"],
              ["1", "tri_down"],
              ["2", "tri_up"],
              ["3", "tri_left"],
              ["4", "tri_right"],
              ["8", "octagon"],
              ["s", "square"],
              ["p", "pentagon"],
              ["P", "plus (filled)"],
              ["*", "star"],
              ["h", "hexagon1"],
              ["H", "hexagon2"],
              ["+", "plus"],
              ["x", "x"],
              ["X", "x (filled)"],
              ["D", "diamond"],
              ["d", "thin_diamond"],
              ["|", "vline"],
              ["_", "hline"],
              [0, "tickleft"],
              [1, "tickright"],
              [2, "tickup"],
              [3, "tickdown"],
              [4, "caretleft"],
              [5, "caretright"],
              [6, "caretup"],
              [7, "caretdown"],
              [8, "caretleft (centered at base)"],
              [9, "caretright (centered at base)"],
              [10, "caretup (centered at base)"],
              [11, "caretdown (centered at base)"],
              ["None", "nothing"],
              ["$sc$", "Render the string using mathtext."]
]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

y=np.random.rand(10,1)
nplots=len(MAT_MARKER)
ncols,ndims=6,3

fig,ax=plt.subplots(
    nrows=math.ceil(nplots/ncols),
    ncols=ncols,
    figsize=(ndims*ncols,
             ndims*nplots/ncols),
    sharex=True,
)

for i,m in enumerate(MAT_MARKER):
  ax[int(i/ncols)][i%ncols].plot(
      y,'o',
      marker=m[0],
      color='purple'
  )
  ax[int(i/ncols)][i%ncols].set_title(str(m[0])+":"+m[1])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.DataFrame({
    'old':np.random.randint(10,size=4),
    'new':np.random.randint(20,size=4),
})
fig,ax=plt.subplots()
for i,row in df.iterrows():
  plt.plot([row.old,row.new],'purple')
plt.xlim(-0.5,1.5)
plt.xticks([0,1])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.xaxis.grid('solid','both',alpha=0.9)
ax.yaxis.grid('dashed','both',alpha=0.1)
plt.title('Slope Graph');
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = list(map(chr,range(65,65+20)))
x = np.random.random(20)
x = [*x,x[0]]
labels_loc = np.linspace(0,2*np.pi,len(x))

plt.subplot(polar=True)
plt.plot(labels_loc,x,'purple')
plt.thetagrids(np.degrees(labels_loc),labels)
plt.title('Radar Plot');
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy.stats.kde import gaussian_kde

X = [norm.rvs(i,2,50) for i in range(5)]
cmap = matplotlib.cm.Purples
normal = matplotlib.colors.Normalize(0,len(X))
colors = [cmap(normal(x)) for x in range(0,len(X))]

X_full = np.concatenate(X)
x = np.linspace(np.min(X_full),np.max(X_full),200)

for i,d in enumerate(X):
  pdf = gaussian_kde(d)
  c = pdf(x)
  y = i*0.1
  plt.fill_between(
      x,np.ones(200)*y,c+y,
      zorder=len(X)-i+1,color=colors[i]
  )
  plt.plot(x,c+y,c='k',zorder=len(X)-i+1)

plt.axis('off')
plt.title('Ridge Plot');
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker

x,y = np.meshgrid(
    np.linspace(1,10,100),
    np.linspace(1,10,10)
)
z = np.exp(x*y)

plt.contourf(x,y,z,
             locator=ticker.LogLocator(),
             cmap="Purples")
plt.title('Contour Plot')
plt.axis('off');
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
n=3
df = pd.DataFrame({
    '0':np.arange(0,n,1),
    '1':np.random.random(size=n),
    '2':np.random.random(size=n),
    '3':np.random.random(size=n),
})
pd.plotting.parallel_coordinates(
    df,'0',lw=3,color='purple'
)
plt.axis('off')
plt.xlim(-0.5,2.5)
plt.legend().remove()
plt.title('Parallel Plot');
```
